# Numerical Integration of growth equation

```python
import numpy as np  # for numerics
import matplotlib.pyplot as plt # to make plots
import seaborn as sns # to make plots more appealing

sns.set(style="darkgrid") # Gray background with white grid
sns.set_context('talk') # Sets default size to the curves and fonts
# display plots next to code
%matplotlib inline 
```

### Variables for numerical integration

```python
# Define parameters for integration
N_0 = 1 # initial number of cells
r = 1 / 1000 # growth rate in s^-1
n_steps = 150 # number of time steps for integration
delta_t = 1 / (100 * r) # time steps in s
```

### Integration

```python
# Initialise array to save number of cells
N_array = np.zeros(n_steps)

# Change initial conditions
N_array[0] = N_0

# For loop to integrate
for i in range(1, n_steps):
    N_array[i] = N_array[i-1] + r * delta_t * N_array[i-1]
```

### Let's Plot

```python
# Plot growth curve
plt.plot(np.arange(n_steps) * delta_t, N_array)

# Label axis
plt.xlabel('time(sec)')
plt.ylabel(r'number of cells $N_0$')
```

### Comparing results with actual results

```python
# Generate fine grid of points
time = np.linspace(0, n_steps * delta_t, 500)

# Apply analytical solution
N_theory = N_0 * np.exp(r * time)
```

### Let's plot them together

```python
# Plot the exact result
plt.plot(time, N_theory, label = 'Theory')

# plot the numerical integration
plt.plot(np.arange(n_steps) * delta_t, N_array,
          linestyle ='--', label = 'Numerical Integration')
# Label axis
plt.xlabel('time(sec)')
plt.ylabel(r'number of cells $N_0$')
plt.legend()
```

## Logistic growth

```python
# Define parameters for integration
N_0 = 1 # initial number of cells
r = 1 / 1000 # growth rate in s^-1
n_steps = 5000 # number of time steps for integration
delta_t = 1 / (100 * r) # time steps in s

K = 10 ** 9 # Growth factor
```

```python
# Initialise array to save number of cells
N_array = np.zeros(n_steps)

# Change initial conditions
N_array[0] = N_0

# For loop to integrate
for i in range(1, n_steps):
    N_array[i] = N_array[i-1] + \
                    r * delta_t * N_array[i-1] * (1 - N_array[i-1] / K)
```

```python
# Plot growth curve
plt.plot(np.arange(n_steps) * delta_t, N_array)

# Label axis
plt.xlabel('time(sec)')
plt.ylabel(r'number of cells $N_0$')
```
